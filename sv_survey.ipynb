As seen from their descriptions, most characters follow fairly common formulas (i.e. the jock, the popular girl, the videogame nerd, the scientist, the poet etc.), therefore I believed it would be interesting to see whether players of different gender, sexual orientation, and age differ in the ranking of these characters. I focused the analysis on the ranking specifically, instead of just allowing players to vote for their favourites, in order to better track changes in their choices and analyse which characters the groups dislike as well as which ones they prefer.

The data was collected by myself through a Google Form survey in July 2018. The form was diffused mainly through the website reddit.com, thus most of the responses are likely to come from the Stardew Valley community of that website. The users would rank the bachelors and the bachelorettes separately, each from #1 to #6, with #1 being the best. Then, the players had the option to choose their 3 favourite characters regardless of gender, and finally to provide information about their gender, orientation and age range.

## Data cleaning

### Timestamp

```python
sv["time"] = pd.to_datetime(sv["time"], format = "%Y/%m/%d %I:%M:%S %p %Z") # convert time of voting into datetime object
sv.head()
```

### Votes

The votes as provided by Google Forms need to be cleaned by removing the "#" and turning the string into an integer.

```python
sv.loc[:,"alex":"penny"] = sv.loc[:,"alex":"penny"].replace("#","", regex=True) # remove "#" from votes
sv.loc[:,"alex":"penny"] = sv.loc[:,"alex":"penny"].apply(pd.to_numeric) # turn into integer
sv.head()
```

As mentioned in the introduction, users ranked the characters from #1 to #6. However, if the ranks were used as is, this would show the highest scoring characters as having the lowest numeric totals. To avoid confusion, the ranks have therefore been converted to points, so that the characters with the most points have the highest votes. The new system will score the candidates from 1 point (lowest) to 6 points (highest).

```python
sv.loc[:,"alex":"penny"] = sv.loc[:,"alex":"penny"].replace({1 : 6, 2 : 5, 3 : 4, 4 : 3, 5 : 2, 6 : 1})
sv.head()
```

### Gender

In order to ensure the voters could choose a gender they most identified with, the following options were provided: "Male", "Female", "Genderqueer/Non-binary", "Prefer not to say", and an open "Other" option. However, the last one turned out to be a terrible idea that will not be replicated in future surveys, as it generated a list of unique responses that cannot be properly classified. Instead, a closed "Other" option is preferable, and the survey has been amended to reflect this.

A list of all unique answers was generated: the answers "Non-binary" and "Genderqueer" were integrated in the provided option "Genderqueer/Non-binary", renominated "Non-binary" for brevity. All the other answers were classified as "Other".

```python
print("Unique values before cleaning:") # list of unique answers
print(sv.loc[:,"gender"].unique())

sv.loc[:,"gender"] = sv.loc[:,"gender"].replace("Genderqueer/Non-binary","Non-binary") # group the three terms under
sv.loc[:,"gender"] = sv.loc[:,"gender"].replace("Genderqueer","Non-binary")            # "Non-binary"

standard_genders = ["Male", "Female", "Non-binary", "Prefer not to say"] # standard answers

gender_bool = ~sv.loc[:,"gender"].isin(standard_genders) # replace all non-standard answers 
sv.loc[gender_bool, "gender"] = "Other"                  # with "Other"

print("Unique values after cleaning:") # new list of unique answers
print(sv.loc[:,"gender"].unique())
```

### Orientation

The same problem that surfaced with the open "Other" option in the previous section presented itself here. In this case, the other options provided were "Asexual", "Homosexual", "Heterosexual", "Bisexual/Pansexual", and "Prefer not to say".

Like above, a list of all unique answers was generated: users who responded "Pansexual" and "Bisexual" were grouped with "Bisexual/Pansexual" users, since, while these identities are very distinct to some people, they are interchangeable for the scope of this survey, as both user groups would experience attraction to male and female candidates. All other answers were classified as "Other".

```python
print("Unique values before cleaning:") # list of unique answers
print(sv.loc[:,"orientation"].unique())

sv.loc[:,"orientation"] = sv.loc[:,"orientation"].replace("Pansexual","Bisexual") # group bisexual + pansexual
sv.loc[:,"orientation"] = sv.loc[:,"orientation"].replace("Bisexual","Bisexual/Pansexual") # rename category

# standard answers
standard_orientations = ["Asexual", "Homosexual", "Heterosexual", "Bisexual/Pansexual", "Prefer not to say"] 

orientation_bool = ~sv.loc[:,"orientation"].isin(standard_orientations) # replace all non-standard answers 
sv.loc[orientation_bool, "orientation"] = "Other"                       # with "Other"

print("Unique values after cleaning:") # new list of unique answers
print(sv.loc[:,"orientation"].unique())
```

### Age

The missing values in the Age column were replaced with a "Prefer not to say" label.

```python
print("Unique values before cleaning:") # list of unique answers
print(sv.loc[:,"age"].unique())

sv.loc[:,"age"].fillna("Prefer not to say", inplace=True) # substitute missing values with "Prefer not to say"

print("Unique values after cleaning:") # new list of unique answers
print(sv.loc[:,"age"].unique())
```

### Average score per candidate
The average score per bachelor/bachelorette calculated across all voters.

```python
total_bachelors = sv.loc[:,"alex":"shane"].sum().sort_values(ascending=False) # sum of votes per candidate
total_avg_bachelors = round(total_bachelors/n_votes, 2) # average score per candidate

print("Average score per bachelor:")
print(total_avg_bachelors)

total_bachelorettes = sv.loc[:,"abigail":"penny"].sum().sort_values(ascending=False)
total_avg_bachelorettes = round(total_bachelorettes/n_votes, 2)

print("Average score per bachelorette:")
print(total_avg_bachelorettes)
```

### Average scores by gender

The average score per bachelor/bachelorette calculated across different genders. The "Prefer not to say" and "Other" options are not included as they do not provide insight into the voter's gender identity.

```python
male_bool = sv.loc[:,"gender"] == "Male" # select group of voters
female_bool = sv.loc[:,"gender"] == "Female"
enby_bool = sv.loc[:,"gender"] == "Non-binary"

bachelors_by_male = calc_points(male_bool, "b") # calculate average scores across group
bachelors_by_fem = calc_points(female_bool, "b")
bachelors_by_enby = calc_points(enby_bool, "b")

bachelors_gender = pd.concat([bachelors_by_male, bachelors_by_fem, bachelors_by_enby], 
                             axis=1, sort=False) # create summary table
bachelors_gender.reset_index(inplace=True)
bachelors_gender.columns = ["bachelor", "men", "women", "non-binary"]

print(bachelors_gender)
```

```python
bachelorettes_by_male = calc_points(male_bool, "be")
bachelorettes_by_fem = calc_points(female_bool, "be")
bachelorettes_by_enby = calc_points(enby_bool, "be")

bachelorettes_gender = pd.concat([bachelorettes_by_male, bachelorettes_by_fem, bachelorettes_by_enby], 
                                 axis=1, sort=False)
bachelorettes_gender.reset_index(inplace=True)
bachelorettes_gender.columns = ["bachelorette", "men", "women", "non-binary"]

print(bachelorettes_gender)
```

### Average scores by orientation

The average score per bachelor/bachelorette calculated across different orientations. The "Prefer not to say" and "Other" options are not included as they do not provide insight into the voter's orientation.

```python
het_bool = sv.loc[:,"orientation"] == "Heterosexual"
bi_bool = sv.loc[:,"orientation"] == "Bisexual/Pansexual"
homo_bool = sv.loc[:,"orientation"] == "Homosexual"
ace_bool = sv.loc[:,"orientation"] == "Asexual"

bachelors_by_het = calc_points(het_bool, "b")
bachelors_by_bi = calc_points(bi_bool, "b")
bachelors_by_homo = calc_points(homo_bool, "b")
bachelors_by_ace = calc_points(ace_bool, "b")

bachelors_orientation = pd.concat([bachelors_by_het, bachelors_by_bi, bachelors_by_homo, bachelors_by_ace], 
                                  axis=1, sort=False)
bachelors_orientation.reset_index(inplace=True)
bachelors_orientation.columns = ["bachelor", "heterosexual", "bisexual", "homosexual", "asexual"]

print(bachelors_orientation)
```

```python
bachelorettes_by_het = calc_points(het_bool, "be")
bachelorettes_by_bi = calc_points(bi_bool, "be")
bachelorettes_by_homo = calc_points(homo_bool, "be")
bachelorettes_by_ace = calc_points(ace_bool, "be")

bachelorettes_orientation = pd.concat([bachelorettes_by_het, bachelorettes_by_bi, 
                                       bachelorettes_by_homo, bachelorettes_by_ace], 
                                       axis=1, sort=False)
bachelorettes_orientation.reset_index(inplace=True)
bachelorettes_orientation.columns = ["bachelorette", "heterosexual", "bisexual", "homosexual", "asexual"]

print(bachelorettes_orientation)
```

### Average scores by gender and orientation

The average score per bachelor/bachelorette calculated by comparing groups that are actually attracted to the gender of the spouses in question, e.g. average bachelor scores for men attracted to men versus women attracted to men. See the corresponding visualisation section for more information.

```python
mlm_bool = (sv.loc[:,"gender"] == "Male") & ((sv.loc[:,"orientation"] == "Homosexual") | 
                                             (sv.loc[:,"orientation"] == "Bisexual/Pansexual"))
bachelors_by_mlm = calc_points(mlm_bool, "b")

wlm_bool = (sv.loc[:,"gender"] == "Female") & ((sv.loc[:,"orientation"] == "Heterosexual") | 
                                               (sv.loc[:,"orientation"] == "Bisexual/Pansexual"))
bachelors_by_wlm = calc_points(wlm_bool, "b")

bachelors_plm = pd.concat([bachelors_by_mlm, bachelors_by_wlm], 
                          axis=1, sort=False)
bachelors_plm.reset_index(inplace=True)
bachelors_plm.columns = ["bachelor", "men who love men", "women who love men"]

print(bachelors_plm)
```

```python
mlw_bool = (sv.loc[:,"gender"] == "Male") & ((sv.loc[:,"orientation"] == "Heterosexual") | 
                                             (sv.loc[:,"orientation"] == "Bisexual/Pansexual"))
bachelorettes_by_mlw = calc_points(mlw_bool, "be")

wlw_bool = (sv.loc[:,"gender"] == "Female") & ((sv.loc[:,"orientation"] == "Homosexual") | 
                                               (sv.loc[:,"orientation"] == "Bisexual/Pansexual"))
bachelorettes_by_wlw = calc_points(wlw_bool, "be")

bachelorettes_plw = pd.concat([bachelorettes_by_mlw, bachelorettes_by_wlw], 
                              axis=1, sort=False)
bachelorettes_plw.reset_index(inplace=True)
bachelorettes_plw.columns = ["bachelorette", "men who love women", "women who love women"]

print(bachelorettes_plw)
```

## Conclusion

The preferences expressed by this sample of Stardew Valley players, while representative only of a fraction of the gaming community, could be useful to those interested in game development to aid the design of non-playable characters. In fact, if different tropes did indeed attract different demographics, this could be used to target specific player groups of interest by creating characters that would appeal to them.

A factor which could have influenced the outcome of this survey is character development, or more specifically the dialogues through which the candidates in the game relate to the player. The characters that players prefer the most, in fact, could simply be those with better developed stories or more interesting dialogue options, altering a possible preference of a demographic for one trope or another. Furthermore, less dedicated players could ignore characters which are more difficult to court in early game, thus never interacting with those enough to rank them positively, or even negatively, again possibly influencing their choices. To carry out a similar study on a larger scale, without the influence of dialogue or in-game mechanics, and independently of any specific community, a survey could be designed to include newly created characters, along with an avatar and a description of their personality, to see whether these results can be replicated outside of the Stardew Valley environment. This would allow to survey player's first impressions of characters, and could possibly be applied to games other than farming/dating simulators.
